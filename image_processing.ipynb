```python
#This code is purposed to process image in roject activity in KHU IoT basic lect
#ure. This code runs in Google Colab.
#Maintainer: Sangjun Park
#Committer: Eunji Lee 

from google.colab.patches import cv2_imshow
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import files

myfile = files.upload()

#Edge detection#################################################################
image = cv2.imread('iot.jpg')
edges = cv2.Canny(image,40,180)
cv2_imshow(edges)
cv2.waitKey()
cv2.destroyAllWindows()

#HSV filter    
img = cv2.imread("iot.jpg")
cv2_imshow(img)

canvas = np.zeros(shape=img.shape, dtype=np.uint8)
canvas.fill(0)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv, (0, 128, 0), (100, 255, 100))

pixcel = 0
for i in range(0,img.shape[0]):
    for j in range(0,img.shape[1]): #image shape   array 
        if mask[i,j]==255: #mask      
            canvas[i,j]=img[i,j]
            pixcel += 1
#canvas= cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB)
print(pixcel)
print(img.shape[0], img.shape[1], img.shape[0] * img.shape[1m])
cv2_imshow(canvas)
cv2.waitKey()
cv2.destroyAllWindows()
```
